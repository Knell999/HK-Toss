```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
```

# 1. 밑 그림 그리기

## 1-1 단일 그래프
 - 캔버스 내에 하나의 Figure, Figure 내에 하나의 plot만 존재하는 경우

```python
import numpy as np
# 데이터 생성

data = np.arange(1, 100)

# 그리기(plot)
plt.plot(data)

# 시각화된 정보 보여주기 - 캔버스가 마무리 됨
plt.show()
```

## 1-2 다중 그래프(multiple graph)
 1개의 Figure 내에 여러 개의 그래프를 그리는 것

```python
data1 = np.arange(1,51)
data2 = np.arange(51,101)

plt.plot(data1)
plt.plot(data2)

plt.show()
```

### [참고] figure()
- `plt.figure()`: 새로운 캔버스 생성

```python
plt.plot(data1) # 그래프를 그리기 전에 figure()가 없으면 자동으로 만들어 준다.

plt.figure()
plt.plot(data2)

plt.show()
```

## 1-3. 한 개의 캔버스 내에 여러 figure 생성

### `subplot`
- 행, 열, 인덱스를 활용해서 그려주는 방식

```python
# subplt(row, column, index)
plt.subplot(2,1,1)
plt.plot(data1)

plt.subplot(2,1,2)
plt.plot(data2)

plt.show()
```

```python
# 콤마 생략 가능
plt.subplot(211)
plt.plot(data1)

plt.subplot(212)
plt.plot(data2)

plt.show()
```

```python
plt.subplot(121)
plt.plot(data1)

plt.subplot(122)
plt.plot(data2)

plt.show()
```

### subplots

```python
# subplots(row, column)
fig, axes = plt.subplots(2,3)

axes[0, 0].plot(data1)
axes[0, 1].plot(data1 * data1)
axes[0, 2].plot(data1 ** 3)

axes[1, 0].plot(data1 ** 4)
axes[1, 1].plot(data1 ** 5)
axes[1, 2].plot(data1 ** 6)
```

```python
fig, axes = plt.subplots(2,3)
data_list = [
    data1,
    data1 ** 2,
    data1 ** 3,
    data1 ** 4,
    data1 ** 5,
    data1 ** 6,
             ]

for plot_data, ax in zip(data_list, axes.ravel()):
  ax.plot(plot_data)

plt.show()
```

## 2-1 title
그래프의 제목

```python
plt.plot(data1)
plt.plot(data2)

# plt.title("여기는 타이틀 입니다.")
plt.title("여기는 타이틀 입니다.", fontsize=20)
plt.show()
```

## 2-2 X, Y축 Label 설정

```python
plt.plot(data1)
plt.plot(data2)

plt.title("label 설정 예제 입니다.")

# x축 label 설정
plt.xlabel("X축", fontsize=20)
plt.ylabel("Y축", fontsize=20)

plt.show()
```

## 2-3 X, Y축 tick 설정 (rotation)

```python
plt.plot(np.arange(10), np.arange(10) * 2)
plt.plot(np.arange(10), np.arange(10) ** 2)
plt.plot(np.arange(10), np.log(np.arange(10)))

# 타이틀 & font 설정
plt.title("X, Y 틱을 조정")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```

## 2-4 범례 ( Legend ) 설정

```python
plt.plot(np.arange(10), np.arange(10) * 2, label="10 * 2")
plt.plot(np.arange(10), np.arange(10) ** 2, label="10 ** 2")
plt.plot(np.arange(10), np.log(np.arange(10)), label="log")

# 타이틀 & font 설정
plt.title("범례 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# legend(범례) 설정
plt.legend(["10 * 2", "10 ** 2", "log"])

plt.show()
```

## 2-5. X와 Y의 한계점 (Limit) 설정
`xlim(), ylim()`

```python
plt.plot(np.arange(10), np.arange(10) * 2)
plt.plot(np.arange(10), np.arange(10) ** 2)
plt.plot(np.arange(10), np.log(np.arange(10)))

# 타이틀 & font 설정
plt.title("x축, y축의 한계점 설정")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

# legend(범례) 설정
plt.legend(["10 * 2", "10 ** 2", "log"])

# x, y의 limit 설정
plt.xlim(0, 5)
plt.ylim(0.5, 10)

plt.show()
```

## 2-6. 스타일 세부 설정 marker, line, color 설정
**marker의 종류**
* '.'	point marker
* ','	pixel marker
* 'o'	circle marker
* 'v'	triangle_down marker
* '^'	triangle_up marker
* '<'	triangle_left marker
* '>'	triangle_right marker
* '1'	tri_down marker
* '2'	tri_up marker
* '3'	tri_left marker
* '4'	tri_right marker
* 's '	square marker
* 'p'	pentagon marker
* '*'	star marker
* 'h'	hexagon1 marker
* 'H'	hexagon2 marker
* '+'	plus marker
* 'x'	x marker
* 'D'	diamond marker
* 'd'	thin_diamond marker
* '|'	vline marker
* '_'	hline marker

```python
plt.plot(np.arange(10), np.arange(10) * 2, marker='o', markersize=5)
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', markersize=10)
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', markersize=15)
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', markersize=20)

# 타이틀 & font 설정
plt.title("마커 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```

**line의 종류**
* '-' solid line style
* '--' dashed line style
* '-.' dash-dot line style
* ':' dotted line style

```python
plt.plot(np.arange(10), np.arange(10) * 2,      marker='o', linestyle='')   # 선이 안보임
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', linestyle='-')  # 실선
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', linestyle='--') # 짧은 실선
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', linestyle='-.') # 선과 점을 엮어서
plt.plot(np.arange(10), np.arange(10) * 2 - 40, marker='v', linestyle=':')  # 점섬

# 타이틀 & font 설정
plt.title("라인 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)
plt.show()
```

**color의 종류**
* 'b'	blue
* 'g'	green
* 'r'	red
* 'c'	cyan
* 'm'	magenta
* 'y'	yellow
* 'k'	black
* 'w'	white

```python
plt.plot(np.arange(10), np.arange(10) * 2,      marker='o', linestyle='', color='b')
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', linestyle='-', color='r')
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', linestyle='--', color='c')
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', linestyle='-.', color='y')

# 타이틀 & font 설정
plt.title("색상 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```

**투명도 설정**
`alpha` 옵션을 0 ~ 1사이로 조절. 값이 작을 수록 투명해 진다.

```python
plt.plot(np.arange(10), np.arange(10) * 2,      marker='o', color='b', alpha=0.1)
plt.plot(np.arange(10), np.arange(10) * 2 - 10, marker='H', color='r', alpha=0.3)
plt.plot(np.arange(10), np.arange(10) * 2 - 20, marker='D', color='c', alpha=0.6)
plt.plot(np.arange(10), np.arange(10) * 2 - 30, marker='*', color='y', alpha=1.0)

# 타이틀 & font 설정
plt.title("투명도( alpha ) 설정 예제")

# X축 & Y축 Label 부여
plt.xlabel("X축")
plt.ylabel("Y축")

# X tick, Y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

plt.show()
```

# 지하철 역 주변 맛집 크롤링

```python
KEYWORD_LOCAL_URL = "https://dapi.kakao.com/v2/local/search/keyword.json?query={}&page={}"

stations = ["성수역", "신촌역", "강남역"]
```

```python
headers = {
    "Authorization": f"KakaoAK {REST_API_KEY}"
}

response = requests.get(KEYWORD_LOCAL_URL.format(stations[0] + " 맛집", 1), headers=headers)
response.json()
```

```python
def scrap_station_local_restaurant(station, page):
  response = requests.get(KEYWORD_LOCAL_URL.format(station + "근처 맛집", page), headers=headers)
  datas = response.json()["documents"]

  df = pd.DataFrame(datas)
  df = df.set_index("id")
  df['station'] = station
  return df

df_list = [ scrap_station_local_restaurant(station, page) for station in stations for page in range(1, 46) ]
```

```python
df_total = pd.concat(df_list)
df_total.head()
```

```python
import requests

kospi_url = "https://api.stock.naver.com/chart/domestic/index/KOSPI?periodType=dayCandle"
```

```python
res = requests.get(kospi_url)
res
```

```python
datas = res.json()['priceInfos']
datas
```

```python
# 데이터 프레임 구성
import pandas as pd

kospi_df = pd.DataFrame(datas)
kospi_df.head()
```

```python
# 코스닥 데이터 가져오기
url = "https://api.stock.naver.com/chart/domestic/index/KOSDAQ?periodType=dayCandle"

res = requests.get(url)
datas = res.json()["priceInfos"]
datas[:2]
```

```python
kosdaq_df = pd.DataFrame(datas)
kosdaq_df.tail()
```

```python
# USD 크롤링
url = "https://m.stock.naver.com/front-api/chart/pricesByPeriod?reutersCode=FX_USDKRW&category=exchange&chartInfoType=marketindex&scriptChartType=areaYear"
```

```python
res = requests.get(url)
datas = res.json()["result"]["priceInfos"]
datas[:2]
```

```python
usd_df = pd.DataFrame(datas)
usd_df.tail()
```

```python
kospi_df.head()
```

```python
kospi_df_0117 = kospi_df[kospi_df["localDate"] >= "20240805"]
kosdaq_df_0117 = kosdaq_df[kosdaq_df["localDate"] >= "20240805"]
usd_df_0117 = usd_df[usd_df["localDate"] >= "20240805"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set() # matplotlib을 활용해도 seaborn 스타일로 사용이 된다.
```

```python
plt.figure(figsize=(20, 10))

plt.plot(kospi_df_0117["localDate"], kospi_df_0117["closePrice"], "r*:", label="kospi")
plt.plot(kosdaq_df_0117["localDate"], kosdaq_df_0117["closePrice"], "bo:", label="kosdaq")
plt.plot(usd_df_0117["localDate"], usd_df_0117["closePrice"], "g^--", label="usd")
plt.xticks(rotation=45, fontsize=7)
plt.legend()
plt.show()
```

```python
# MinMax Scaling
from sklearn.preprocessing import minmax_scale

plt.figure(figsize=(20, 8))

plt.plot(kospi_df_0117["localDate"], minmax_scale(kospi_df_0117["closePrice"]), "r*:", label="kospi")
plt.plot(kosdaq_df_0117["localDate"],minmax_scale(kosdaq_df_0117["closePrice"]), "bo:", label="kosdaq")
plt.plot(usd_df_0117["localDate"],   minmax_scale(usd_df_0117["closePrice"]), "g^--", label="usd")
plt.xticks(rotation=90, fontsize=7)
plt.legend()
plt.show()
```

```python
# 코스피, 코스닥, 미 환율 상관관계 확인
import numpy as np

# 코스피 : 코스닥 상관관계
np.corrcoef(kospi_df_0117["closePrice"], kosdaq_df_0117["closePrice"])[0, 1]
```

```python
# 코스피 : 미 달러 환율 상관관계
np.corrcoef(kospi_df_0117["closePrice"], usd_df_0117["closePrice"])[0, 1]
```

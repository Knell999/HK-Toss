```python
html_str = """
<html>
  <head>
    <title>안녕하세요</title>
  </head>
  <body>
    <div id="container">
      <p class='p1'>hello</p>
      <p>Bye</p>
    </div>
  </body>
</html>"""
```

html 문자열을 파이썬에서 다뤄야 한다.
- HTML 문자열을 엘리먼트화 해서 파이썬에서 **객체**로 다룰 수 있어야 한다.
  - `BeautifulSoup`을 이용해서 HTML 문자열을 엘리먼트 객체로 바꿔줄 수 있다!

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_str, 'html.parser')
soup
```

```python
type(soup)
```

# find 방식
- `find("태그명", "속성값 딕셔너리")`
  - 한 개의 Element만 찾기
- `find_all("태그명", "속성값 딕셔너리")`
  - 여러 개의 Element 찾기(리스트)

```python
# div 중, id 속성이 container인 것 찾기
soup.find("div", {"id": "container"})
```

```python
# p 중 클래스가 p1인 것 찾기
soup.find_all("p", {"class" : "p1"})
```

```python
soup.find_all("p")
```

```python
# 제일 마지막 p 엘리먼트 가져오기
soup.find_all("p")[-1]
```

```python
p_p1 = soup.find("p", {"class" : "p1"})
p_p1
```

```python
# 동생 찾기 : nextSibling
p_p1.nextSibling.nextSibling
```

```python
# 텍스트 찾기
p_p1.text
```

```python
p_p1.get_text()
```

# 선택자(selector) 사용 방식
- `soup.select(선택자)` : 선택자에 의해 엘리먼트를 **여러 개** 선택
- `soup.select_one(선택자)` : 선택자에 의해 엘리먼트를 한 개만 선택

```python
soup.select_one("#container")
```

```python
soup.select("#container > .p1")
```

# 네이버 환율정보 가져오기

```python
url = "https://finance.naver.com/marketindex/"
```

```python
import requests
from bs4 import BeautifulSoup

response = requests.get(url)
page = response.content # content : 접속한 웹 서버의 응답을 문자열로 가져오는 속성
page
```

```python
soup = BeautifulSoup(page, 'html.parser')
soup
```

```python
exchange_list = soup.select('#exchangeList > li')
exchange_list
```

```python
test_fin = exchange_list[0]
test_fin
```

```python
# 대상화폐
test_fin.select_one("h3.h_lst").text
```

```python
float(test_fin.select_one("span.value").text.replace(',', ''))
```

```python
float(test_fin.select_one("span.change").text.replace(',', ''))
```

```python
test_fin.select("span.blind")[-1].text
```

```python
# 전체 데이터 스크래핑
c_name_list = []
exchange_rate_list = []
change_list = []
updown_list = []

for exchange in exchange_list:
  c_name = exchange.select_one("h3.h_lst").text
  exchange_rate = float(exchange.select_one("span.value").text.replace(',', ''))
  change = float(exchange.select_one("span.change").text.replace(',', ''))
  updown = exchange.select("span.blind")[-1].text

  print(c_name, exchange_rate, change, updown)

  c_name_list.append(c_name)
  exchange_rate_list.append(exchange_rate)
  change_list.append(change)
  updown_list.append(updown)
```

```python
import pandas as pd

exchange_datas = {
    "국가" : c_name_list,
    "환율" : exchange_rate_list,
    "변동" : change_list,
    "등락" : updown_list
}

df_exchange = pd.DataFrame(exchange_datas)
df_exchange
```

```python
df_exchange.to_csv("./20250121_환율정보.csv", index=False)
```

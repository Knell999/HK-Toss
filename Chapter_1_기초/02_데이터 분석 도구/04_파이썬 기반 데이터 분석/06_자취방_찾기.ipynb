```python
data = data.query('월세 > 0')
data['보증금'].unique()
```

```python
data = data.dropna(subset=['보증금'])
data.isnull().sum()
```

```python
# 보증금 3000 이하 자취방을 찾을 계획이니 보증금에 '억'이 들어가면 제외
# not : ~로 표시
# data = data.query('~보증금.str.contains("억")')
data = data.query('~보증금.str.contains("억")')
data['보증금'].unique()
```

```python
data['보증금'] = data['보증금'].str.replace(',', '').astype(int)
data['보증금'].unique()
```

```python
# 물건층 / 전체층 분리 후 1층과 꼭대기층 유무 컬럼 만들기
# 지하의 경우 음수로 표시
data.head()
```

```python
data[['물건층', '전체층']] = data['층수(물건층/전체층)'].str.split("/", expand=True)
data.head()
```

```python
data['물건층'].unique()
```

```python
data['전체층'].unique()
```

```python
# 제외의 기준 : 숫자가 아닌, 저, 중, 고, B(지하층)이 있으면 제외
# 전체층과 물건층이 같으면 제외(꼭대기층)
# 1층 제외

"B1".isnumeric() # 문자열이 숫자로만 이루어져 있는가?를 확인하는 메소드
```

```python
def floor_info(target, total):
  '''
    floor_info : 비선호층이면 "y", 선호층이면 "n"
    target : 물건층
    total : 전체층
  '''
  try:
    if not target.isnumeric(): # 저, 중, 고, B1 등이면 'Y'
      return "y"

    elif int(target) == 1 or int(target) /int(total) == 1:
      return "y"

    else:
      return "n"
  except:
    return "n"

data ['비선호층여부'] = data.apply(lambda x : floor_info(x['물건층'], x['전체층']), axis=1)
```

```python
data.head()
```

```python
# 보증금 3000만원 이하
# 지하, 반지하, 꼭대기층 ,저 , 중, 고 애매한 층은 배제
# 북향 비선호
data_filtered = \
  data.query('300 <= 보증금 <= 3000 and 비선호층여부 == "n" and 전체층 != "1" and ~방향.str.contains("북")')
data_filtered.head()
```

```python
# 태그 컬럼 분리
data_filtered['태그']
```

```python
import ast

test_tags = data_filtered['태그'][1]
ast.literal_eval(test_tags)
```

```python
import numpy as np

max_columns = 4
tags_df = pd.DataFrame(
    data_filtered['태그']
    .apply(lambda x : ast.literal_eval(x))
    .apply(lambda x : x + [np.nan] * (max_columns - len(x)) if len(x) < max_columns else x).to_list(),
    columns=[
        "tag1", "tag2", "tag3", "tag4"
    ]
)

tags_df.index = data_filtered.index
tags_df.head()
```

```python
data_filtered = pd.concat([data_filtered, tags_df], axis=1)
data_filtered.head()
```

```python
data_filtered = data_filtered.query("tag1.str.contains('년')") # 연식 정보가 있는 데이터만 필터링
data_filtered['연식'] = [int(data[0]) for data in data_filtered['tag1'].str.split('년')]
data_filtered.head()
```

```python
data_filtered = data_filtered[['물건번호','월세','보증금','전용면적(m2)','방향','위도','경도','물건층','전체층','연식']]
data_filtered.head()
```

```python
from haversine import haversine

haversine((37.487462,126.913149), (37.484201,126.929715), unit = 'm')
```

```python
def distance(statation_name, lat, lng):
  station_lat = coordinate.query(f'역사명 == "{statation_name}"')['위도'].values[0]
  station_lng = coordinate.query(f'역사명 == "{statation_name}"')['경도'].values[0]

  distance = haversine((station_lat, station_lng), (lat, lng), unit='m')

  return distance
```

```python
for s in station_list:
  data_filtered[s] = data_filtered.apply( lambda x : distance(s, x['위도'], x['경도']), axis=1)

data_filtered.head()
```

```python
data_filtered['역까지최소거리'] = data_filtered.apply(
    lambda x : min([x['신대방'], x['신림'], x['봉천'], x['서울대입구(관악구청)'], x['낙성대'], x['사당']]), axis=1
)
```
